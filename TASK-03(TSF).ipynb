```python
dataset = dataset.drop(['Postal Code'], axis=1)
dataset.head()
```

```python
# Exploratory Data Analysis

# sns.pairplot(,kind="kde")
sns.pairplot(dataset, corner=True)
```

```python
dataset.corr()
plt.figure(figsize=(10,5))
sns.heatmap(dataset.corr(), annot=True,cmap="PiYG")
plt.show()
```

```python
plt.figure(figsize=(6,6))
textprops = {'fontsize':15}
plt.title('Category')
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(dataset['Category'].value_counts(), labels=dataset['Category'].value_counts().index,autopct='%1.1f%%',textprops = textprops, colors=colors)
plt.show()
```

```python
plt.figure(figsize = (10,16))
dataset.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.ylabel('Profit')
plt.show()
```

```python
# Computing top categories in terms of sales and profit from first 100 observation and plotting them.

top_category_s = dataset.groupby('Category').Sales.sum().nlargest(n=100)
top_category_p = dataset.groupby('Category').Profit.sum().nlargest(n=100)
plt.style.use('seaborn')
top_category_s.plot(kind='bar', figsize=(10,5), fontsize=14)
top_category_p.plot(kind='bar', figsize=(10,5), fontsize=14, color='black')
plt.xlabel('Category', fontsize=15)
plt.ylabel('Total Sales/Profits', fontsize=15)
plt.title('Top Category Sales vs Profit', fontsize=15)
plt.show()
```

```python
# Computing top sub-categories in terms of sales and profit from first 100 observation and plotting them.

top_subcategory_s = dataset.groupby('Sub-Category').Sales.sum().nlargest(n=100)
top_subcategory_p = dataset.groupby('Sub-Category').Profit.sum().nlargest(n=100)
plt.style.use('seaborn')
top_subcategory_s.plot(kind='bar', figsize=(10,5), fontsize=14)
top_subcategory_p.plot(kind='bar', figsize=(10,5), fontsize=14, color='green')
plt.xlabel('Sub-Category', fontsize=15)
plt.ylabel('Total Sales/Profits', fontsize=15)
plt.title('Top Sub-Category Sales vs Profit', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(14,12))
statewise = dataset.groupby(['Sub-Category'])['Profit'].sum().nlargest(50)
statewise.plot.barh()
```

```python
plt.figure(figsize=(8,7))
sns.lineplot(dataset['Discount'], dataset['Profit'], data=dataset,color="green")
```

```python
plt.figure(figsize = (6,6))
plt.title('Region')
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(dataset['Region'].value_counts(), colors=colors,labels=dataset['Region'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

```python
# The graph above shows that West and East have same profit though sales in the East are less as compared to West.
# Visualising the Sales vs Profits in different states

top_states_s = dataset.groupby("State").Sales.sum().nlargest(n=10)
top_states_p = dataset.groupby("State").Profit.sum().nlargest(n = 10)
plt.style.use('seaborn')
top_states_s.plot(kind = 'bar',figsize = (10,5),fontsize = 14)
top_states_p.plot(kind = 'bar',figsize = (10,5),fontsize = 14, color = 'yellow')
plt.xlabel('States',fontsize = 15)
plt.ylabel('Total sales',fontsize = 15)
plt.title("Top 10 states Sales vs Profit",fontsize = 15)
plt.show()
```

```python
plt.style.use('seaborn')
dataset.plot(kind = "scatter",figsize = (10,5), x = "Sales", y= "Profit", c = "Discount", s = 20,fontsize = 16, colormap = "rainbow")
plt.ylabel('Total Profits',fontsize = 16)
plt.title("Interdependency of Sales,Profits and Discounts",fontsize = 16)
plt.show()
```

The graph clearly shows that if we giive more Discount on our products sales increases but profit decreases.
